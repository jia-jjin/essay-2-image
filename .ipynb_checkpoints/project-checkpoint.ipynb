# essay-2-image backend

## Importing the data

Combining all data sources into a single df/csv for later reference

```python
import numpy as np
import pandas as pd

data1 = np.load('data/COCO_test_1k.npy')
data2 = np.load('data/diffusiondb_test_1k.npy')
data3 = np.load('data/lexica_test_1k.npy')
data4 = pd.read_csv("data/1k.csv")['prompt'].values
df = pd.DataFrame({
    "prompt": np.concatenate((data1.flatten(), data2.flatten(), data3.flatten(), data4)) 
})
```

```python
df.to_csv('raw.csv', index=False)
```

```python
df.head()
```

## Data preprocessing

### Removing null values

```python
df.isna().sum()
```

```python
df = df.dropna()
```

### Removing duplicated values

```python
df.duplicated().value_counts()
```

```python
df = df.drop_duplicates()
```

### Removing stopwords & punctuations

```python
import nltk
import string
from nltk.corpus import reuters, stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download(['reuters','punkt','stopwords'])
```

```python
stop_words = stopwords.words('english')
```

```python
df['prompt'] = df['prompt'].apply(lambda row: ''.join(list(filter(lambda char: char not in string.punctuation, row))))
```

```python
df['prompt'] = df['prompt'].apply(lambda row: ' '.join(list(filter(lambda char: char not in stop_words and char != "", row.lower().split(' ')))))
```

```python
df.head()
```

## NLP Analysis

```python
words = word_tokenize(' '.join(df.values.flatten()))
```

```python
print(f"Word count: {len(words)}")
print(f"Average word length: {round(sum(list(map(len, words)))/len(words),2)}")
print(f"Vocabulary richness: {round(len(list(set(words)))/len(words)*100,2)}%")
```

```python
from nltk import pos_tag, ne_chunk
nltk.download(['averaged_perceptron_tagger_eng','maxent_ne_chunker','maxent_ne_chunker_tab','words','wordnet'])
```

labelling words by tags

```python
tagged_words = pos_tag(words)
print(tagged_words,flush=True)
```

```python
ne_tree = ne_chunk(tagged_words)
for chunk in ne_tree:
    if hasattr(chunk,'label'):
        print(chunk.label(), ':',' '.join(c[0] for c in chunk))
```

```python
from nltk.probability import FreqDist
import matplotlib.pyplot as plt
```

creating frequency distribution

```python
freq_dist = FreqDist(words)
```

## Wordcloud

```python
from wordcloud import WordCloud
wc = WordCloud(background_color='white').generate(' '.join(words))
plt.figure(figsize=(100,60))
plt.imshow(wc)
plt.axis('off')
plt.savefig(f'wordcloud.png',
            dpi = 200)
```

## Words Frequency Distribution

```python
plt.figure(figsize=(12,6))
freq_dist.plot(50)
plt.savefig(f'freqdist.png',
            dpi = 200)
```

saving cleaned data to a csv

```python
df.to_csv("cleaned.csv", index=False)
```

# Documentation and Assumptions

Based on the wordcloud and words' frequency distribution graph generated, we can see that most occured words are the ones that describe the image generated, e.g. art, potrait, detailed, painting, concept, artstation etc. 

This means that based on the sentence/essay we are going to summarize, we must add a few keywords that describes the image generated for it to fit the general theme better.

E.g. art for cartoony images, potrait when there's only a few characters involved in the image, painting when it shows an event from the past.
